# Load Data from Files

```python
# Generate a list of file names
import glob
file_names = []
for file in glob.glob('Data/NYT/*/*.rtf'):
    file_names.append(file)

scmp_names= []
for file in glob.glob('Data/SCMP/*/*.rtf'):
    scmp_names.append(file)

print(file_names[0])
```

```python
from striprtf.striprtf import rtf_to_text
def read_rtf_to_text(file_name):
    '''
    read one file by its name
    :param file_name:
    :return: plain text
    '''
    with open(file_name, 'r') as file:
        file_text = file.read()
        text = rtf_to_text(file_text, errors="ignore")
    return text
```

```python
#read all files, save in list with plain text
articles_list = []
for name in file_names:
    article = read_rtf_to_text(name)
    articles_list.append(article)
```

```python
test_article = articles_list[1]
print(test_article)
```

```python
import pandas as pd
  
# create an Empty pandas DataFrame with column names
df = pd.DataFrame(columns = ['source', 'date', 'title', 'words', 'body']) #ignoring author since some articles have no author specified
print(df)

#reading all files, saving them in dataframe
articles_list = []
for name in file_names:
    article = read_rtf_to_text(name)
    articles_list.append(article)
    print(name)
    split = article.split("\n\n")
    article_dict = {}
    article_dict["source"] = split[2].split("\n")[0]
    article_dict["date"] = split[2].split("\n")[1]
    article_dict["title"] = name.split("/")[-1][:-4]
    article_dict["words"] = split[3].split("\n")[3].split("\xa0")[1].partition('words')[0] # since len of body results in number of letters
    article_dict["body"] = split[4]
    df = df.append(article_dict, ignore_index=True)    

```
